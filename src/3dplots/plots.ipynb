```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import pickle
import matplotlib
```

```python
# Load dataset
def load_data(plot_dir, label_dir):
    with open(plot_dir, 'rb') as f:
        plot = pickle.load(f)
    with open(label_dir, 'rb') as f:
        label = pickle.load(f)
    return plot, label
```

```python
# betaVAE
plot_dir = r'./plot_3d_beta.pickle'
label_dir = r'./plot_labels_beta.pickle'
plot, label = load_data(plot_dir, label_dir)

colors = ['red', 'blue', 'green']
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(plot[:,0], plot[:,1], plot[:,2], c=label, cmap=matplotlib.colors.ListedColormap(colors))
ax.view_init(-170,60)

ax.set_xlabel('x-pos', fontsize=20, rotation=150)
ax.set_ylabel('y-pos', fontsize=20)
ax.set_zlabel('shape', fontsize=20, rotation=60)

plt.show()
plt.close('all')
```
